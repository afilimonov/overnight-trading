```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
```

```python
def returns(ticker, ax, period='max'):
    history = yf.Ticker(ticker).history(period=period) 
    price_open = history['Open'][0]
    # print(price_open)

    history['close_to_close_return'] = history['Close'].pct_change()
    history['close_to_open_return'] = history['Open']/history['Close'].shift() - 1
    history['open_to_close_return'] = history['Close']/history['Open'] - 1
    history['buy_and_hold'] = history['Close']/price_open - 1
 
    history['close_to_close_return'].cumsum().plot(label="Close to Close", ax=ax)
    history['close_to_open_return'].cumsum().plot(label="Close to Open (overnight)", ax=ax)
    history['open_to_close_return'].cumsum().plot(label="Open to Close (intraday)", ax=ax)
    history['buy_and_hold'].plot(label='Buy and Hold', ax=ax)

    ax.set_title('{} - {}'.format(ticker, period))
    ax.set_xlabel("")    
```

```python
periods = ['ytd', '1y', '2y', '5y', '10y']
tickers = ['AAPL','MSFT','UNH','O','T', 'MMM', 'DE', 'MCD', 'JNJ', 'PFE']
fig, axs = plt.subplots(nrows=len(tickers), ncols=len(periods), figsize=(5*len(periods), 5*len(tickers)))
plt.subplots_adjust(hspace=0.5)

row = 0
for ticker in tickers: 
    col = 0
    for period in periods:
        returns(ticker, axs[row, col], period)
        col += 1
    row += 1

handles, labels = axs[0, 0].get_legend_handles_labels()
l = fig.legend(handles, labels, loc='upper center')
# plt.savefig('output.png')

```

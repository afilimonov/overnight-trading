```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
```

```python
prices_yf = yf.Ticker("O")
prices_yf_hist = prices_yf.history(period="10y")
```

```python
# calc them
price_open = prices_yf_hist['Open'][0]

prices_yf_hist['close_to_close_return'] = prices_yf_hist['Close'].pct_change()
prices_yf_hist['close_to_open_return'] = prices_yf_hist['Open']/prices_yf_hist['Close'].shift() - 1
prices_yf_hist['open_to_close_return'] = prices_yf_hist['Close']/prices_yf_hist['Open'] - 1
prices_yf_hist['buy_and_hold'] = (prices_yf_hist['Close'] - price_open)/price_open
```

```python
# plot them
prices_yf_hist['close_to_close_return'].cumsum().plot(label="Close to Close")
prices_yf_hist['close_to_open_return'].cumsum().plot(label="Close to Open (overnight)")
prices_yf_hist['open_to_close_return'].cumsum().plot(label="Open to Close (intraday)")
prices_yf_hist['buy_and_hold'].plot(label='Buy and Hold')

plt.legend()
```
